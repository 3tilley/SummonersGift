```python
div = pd.read_sql_table("Division", eDb)
lan = pd.read_sql_table("Lane", eDb)
reg = pd.read_sql_table("Region", eDb)
rol = pd.read_sql_table("Role", eDb)
tier = pd.read_sql_table("Tier", eDb)
```

```python
dffinal = dfr[dfr.isStandardLanesGame].copy().merge(df3, "left", on=["matchId", "participantId"] ).rename(columns={"championId": "ChampionId"})
dffinal["role"] = np.where(dffinal.lane=="JUNGLE", dffinal.lane, dffinal.role)
```

```python
# Full split

dfg = dffinal.groupby(["tier","division","ChampionId", "role","StatId"])
dfagg = dfg["statValue"].agg([np.mean, 'count', np.std, 'sem'])
dfagg2 = dfagg.reset_index()


dfagg3 = dfagg2.merge(div,"left",left_on="division",right_on="DivisionJson") \
                  .merge(tier,"left",left_on="tier",right_on="TierJson") \
                  .merge(rol,"left",left_on="role",right_on="RoleJson")
        
dfaggAll = dfagg3[["TierId","DivisionId", "RoleId", "ChampionId", "StatId", "mean", "count", "std", "sem"]]
dfaggAll.head()
```

```python
# Split by champion

dfg = dffinal.groupby(["ChampionId", "role","StatId"])
dfagg = dfg["statValue"].agg([np.mean, 'count', np.std, 'sem'])
dfagg2 = dfagg.reset_index()

dfagg3 = dfagg2.merge(rol,"left",left_on="role",right_on="RoleJson")
        
dfagg3["TierId"],dfagg3["DivisionId"] = 0,0
            
dfaggChamp = dfagg3[["TierId","DivisionId", "RoleId", "ChampionId", "StatId", "mean", "count", "std", "sem"]]
dfaggChamp.head()
```

```python
# Split by champion and tier

dfg = dffinal.groupby(["ChampionId", "role","StatId","tier"])
dfagg = dfg["statValue"].agg([np.mean, 'count', np.std, 'sem'])
dfagg2 = dfagg.reset_index()

dfagg3 = dfagg2.merge(rol,"left",left_on="role",right_on="RoleJson").merge(tier,"left",left_on="tier",right_on="TierJson")
        
dfagg3["DivisionId"] = 0
            
dfaggChampTier = dfagg3[["TierId","DivisionId", "RoleId", "ChampionId", "StatId", "mean", "count", "std", "sem"]]
dfaggChampTier.head()
```

```python
# Full split

dfg = dffinal.groupby(["tier", "division", "role", "StatId"])
dfagg = dfg["statValue"].agg([np.mean, 'count', np.std, 'sem'])
dfagg2 = dfagg.reset_index()


dfagg3 = dfagg2.merge(div,"left",left_on="division",right_on="DivisionJson") \
                  .merge(tier,"left",left_on="tier",right_on="TierJson") \
                  .merge(rol,"left",left_on="role",right_on="RoleJson")

dfagg3["ChampionId"] = 0
        
dfaggRole = dfagg3[["TierId","DivisionId", "RoleId", "ChampionId", "StatId", "mean", "count", "std", "sem"]]
dfaggRole.head()
```

```python
# Full split

dfg = dffinal.groupby(["tier", "division", "StatId"])
dfagg = dfg["statValue"].agg([np.mean, 'count', np.std, 'sem'])
dfagg2 = dfagg.reset_index()


dfagg3 = dfagg2.merge(div,"left",left_on="division",right_on="DivisionJson") \
                  .merge(tier,"left",left_on="tier",right_on="TierJson")

dfagg3["ChampionId"], dfagg3["RoleId"] = 0, 0
        
dfaggRank = dfagg3[["TierId","DivisionId", "RoleId", "ChampionId", "StatId", "mean", "count", "std", "sem"]]
dfaggRank
```

```python
dfCon = pd.concat([dfaggAll, dfaggChamp, dfaggChampTier, dfaggRole, dfaggRank][::-1])
dfCon["MapId"], dfCon["RegionId"], dfCon["DataVersion"], dfCon["Minute"] = 11, 3, 1, 10
dfCon.head(20)
```
