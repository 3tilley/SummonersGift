```python
# Strip out games where there were not standard lanes.
a = df2[["matchId", "teamId","lane", "role"]].copy()
a["laneAndRole"] = a.lane + "_" + a.role
gs = a.groupby(["matchId", "teamId"])#.sum()
rs = gs.aggregate(lambda s: " ".join(list(np.sort(s))))["laneAndRole"].reset_index()
rs["isStandardLanesTeam"] = np.equal(rs.laneAndRole, "BOTTOM_DUO_CARRY BOTTOM_DUO_SUPPORT JUNGLE_NONE MIDDLE_SOLO TOP_SOLO")
#rs["isStandardLanesGame"] = rs.groupby("matchId")["isStandardLanesTeam"].all(True)
#rs["isStandardLanesGame"] = rs.groupby("matchId")["isStandardLanesTeam"].aggregate(np.alltrue)
rs["isStandardLanesGame"] = rs.groupby("matchId")["isStandardLanesTeam"].transform(np.alltrue)
rs = rs.drop("laneAndRole", axis=1)
rs.head()
```

```python
#goodMatches = rs[(rs.laneAndRole == "BOTTOM_DUO_CARRY BOTTOM_DUO_SUPPORT JUNGLE_NONE MIDDLE_SOLO TOP_SOLO")].drop("laneAndRole", axis=1)
df3 = df2.merge(rs, "inner", left_on=["matchId", "teamId"], right_on=["matchId", "teamId"])
df3["csAt10"] = df3.timeline.apply(lambda x:  x.get("creepsPerMinDeltas", {"zeroToTen": None})["zeroToTen"])
df3["csAt10"] = df3["csAt10"] * 10.0

dataLoading.sumCats(df3)

dfc = df3[(df3.role.isin(["SOLO","DUO_CARRY"])) & (df3.isStandardLanesGame)]
    
dfc.head()
```

```python

dfn["splt"] = dfn.name.apply(lambda s: s.split(" "))
dfn["part1"], dfn["part2"] = zip(*dfn.splt.apply(lambda s: (" ".join(s[:-2])," ".join(s[-2:])) if s[-2:]==["Witch", "Hunters"] else (" ".join(s[:-1]),s[-1])))
dfn.head()
```
