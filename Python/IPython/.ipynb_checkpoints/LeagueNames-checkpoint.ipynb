```python
df = dataLoading.loadSummoners(summonerPath)
df.head()
```

```python
nms = df[["name", "tier"]].drop_duplicates()
splt = nms.name.apply(lambda s: s.split(" "))
dfn = pd.DataFrame({"name":nms.name, "tier": nms.tier, "words":splt.apply(len)})
dfn.sort("words",ascending=False, inplace=True)
dfn["splt"] = dfn.name.apply(lambda s: s.split(" "))
dfn["part1"], dfn["part2"] = zip(*dfn.splt.apply(lambda s: (" ".join(s[:-2])," ".join(s[-2:])) if s[-2:]==["Witch", "Hunters"] else (" ".join(s[:-1]),s[-1])))
dfn.head()
```

```python
dfn.groupby(["name", "tier"]).apply(np.shape).reset_index().drop([0], axis=1)#.pivot(columns="tier")
```

```python
dfn2 = dfn[["part1", "part2", "tier"]].copy()

dfn2["part1_prop"] = (dfn2.groupby(["part1","tier"]).transform(len) / len(dfn2))

dfn2["part2_prop"] = (dfn2.groupby(["part2", "tier"])["part1"].transform(len) / len(dfn2))

dfn2.sort(["part2_prop","part1_prop"], ascending=True, inplace=True)
dfn2["part1_num"] = pd.factorize(dfn2.part1)[0]
dfn2["part2_num"] = pd.factorize(dfn2.part2)[0]

cols = {"BRONZE": ("orange", 20),
        "SILVER": ("0.8", 20),
        "GOLD": ("yellow", 20),
        "PLATINUM": ("0.3", 50),
        "DIAMOND": ("green", 100),
        "MASTER": ("blue", 250),
        "CHALLENGER": ("red", 250)}

dfn2["colour"], dfn2["radius"] = zip(*dfn2.tier.apply(cols.get))
dfn2.sort("radius", ascending=False, inplace=True)

dfn2.plot(x="part1_num", y="part2_num", kind="scatter", s=dfn2.radius, c=dfn2.colour, figsize=(15,15), edgecolor='none', legend=True)

#(str(dfn2.tier) + " - " + dfn2.colour).value_counts(dropna=False)
dfn2[(dfn2.radius==100)]
#dfn2.head()
#a = matplotlib.colors.ColorConverter
#a.to_rgb("red")
```
