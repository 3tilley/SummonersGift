```python
dfr = df[(df.queueType=="RANKED_SOLO_5x5")].copy()
dfr.head()
```

```python
matchesToLoad = dfr.matchId
columnsToMakeCategories = ["region", "matchType", "platformId", "matchMode", "matchVersion", "season", "queueType"]

def getFiles(ids):
    dataDict = []
    for f in ids:
                with open(matchesPath + "\\" + str(f) + ".json") as fp:
                    jsonObj = json.load(fp)
                    jsonObj.pop("timeline")
                    if jsonObj["queueType"] not in ["RANKED_SOLO_5x5", "NORMAL_5x5_BLIND"]:
                        jsonObj["participantIdentities"] = None
                        jsonObj["participants"] = None
                    dataDict.append(jsonObj)
                    fp.close()
    df = pd.DataFrame(dataDict)
    for c in columnsToMakeCategories:
        df[c] = df[c].astype("category")
    del dataDict[:]
    return df
%time dfr2 = getFiles(matchesToLoad)
#142652
```

```python
def removeSummonerName(jsonObj):
    for i in jsonObj:
        del i["player"]["summonerName"]
        return i

dfr2.participantIdentities = dfr2.participantIdentities.apply(removeSummonerName)
dfr2.head()
```

```python
summonerInfo = {"matchId":[], "highestAchievedSeasonTier":[], "participantId":[], "data":[]}
summonerIds = {"matchId":[], "participantId":[], "summonerId":[]}
def addToSummonerInfo(sx):
    s = sx["participants"]
    ids = sx["participantIdentities"]
    for i in ids:
        summonerIds["matchId"].append(sx["matchId"])
        
        summonerIds["participantId"].append(i["participantId"])
        summonerIds["summonerId"].append(i["player"]["summonerId"])
    for j in s:
        summonerInfo["matchId"].append(sx["matchId"])
        summonerInfo["highestAchievedSeasonTier"].append(j["highestAchievedSeasonTier"])
        summonerInfo["participantId"].append(j["participantId"])
        summonerInfo["data"].append(j)
        
for i, r in dfr2.iterrows():
    addToSummonerInfo(r)
```

```python
dfs = pd.DataFrame(summonerInfo)
dfs2 = pd.DataFrame(summonerIds)
dfs = dfs.merge(dfs2, how="left",on=["matchId","participantId"])
dfs.head(20)
```

```python
# Load summoners
summonersPath = r'C:\Users\Max\Documents\GitHub\SummonersGift\Data\Summoners\AprilData'
files = os.listdir(summonersPath)

summonerColumnsToMakeCategories = [("division", ["I","II","III","IV","V"][::-1]),
                                   ("tier", ["BRONZE", "SILVER","GOLD","PLATINUM","DIAMOND","MASTER","CHALLENGER"])]

def getFiles(n):
    dataDict = []
    for f in files[:n]:
                with open(summonersPath + "\\" + f) as fp:
                    jsonObj = json.load(fp)
                    for i,v in jsonObj.items():
                        for j in v:
                            if j["queue"] == "RANKED_SOLO_5x5":
                                lst = {"tier":j["tier"], "name":j["name"]}
                                lst.update(j["entries"][0])
                                dataDict.append(lst)
                    fp.close()
    df = pd.DataFrame(dataDict).convert_objects(convert_numeric=True)
    for c,o in summonerColumnsToMakeCategories:
        df[c] = df[c].astype("category", categories=o, ordered=True)
        #df[c].cat.categories = sorted(list(df[c].cat.categories),key=lambda x: o.index(x))
    del dataDict[:]
    return df.drop("miniSeries", axis=1)
%time sums = getFiles(-1)
sums.head()
```
